```python
import numpy as np 
import matplotlib.pyplot as plt 

n_data = 300 # 데이터를 300개 찍는다. 

rng = np.random.default_rng(0) 
x = rng.normal(size=n_data) # 정규분포 normal distribution
y = rng.normal(size=n_data)

plt.scatter(x,y)
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
n_data = 300

rng = np.random.default_rng(0)
X = rng.normal(size=(n_data,2)) # 300개 점인데, 2차원이니까 (x,y)형태

plt.scatter(X[:,0],X[:,1]) # 첫번째 column , 두번쨰 column
plt.show()
```
