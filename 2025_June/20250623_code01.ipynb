```python
import numpy as np 
import matplotlib.pyplot as plt 

n_data = 300 #  300 . 

rng = np.random.default_rng(0) 
x = rng.normal(size=n_data) #  normal distribution
y = rng.normal(size=n_data)

plt.scatter(x,y)
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
n_data = 300

rng = np.random.default_rng(0)
X = rng.normal(size=(n_data,2)) # 300 , 2 (x,y)

plt.scatter(X[:,0],X[:,1]) #  column ,  column
plt.show()
```
