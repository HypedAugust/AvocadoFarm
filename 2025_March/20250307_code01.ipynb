```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self, value):
        self.head = Node(value)

    def append(self, value):
        cur = self.head
        while cur.next is not None:
            cur = cur.next
        cur.next = Node(value)


def get_single_linked_list_sum(linked_list):
    sum = 0 
    cur = linked_list.head 
    while cur is not None:
        sum = sum*10 + cur.data
        cur = cur.next
    return sum 

def get_linked_list_sum(linked_list1, linked_list2):
    sum1 = get_single_linked_list_sum(linked_list1)
    sum2 = get_single_linked_list_sum(linked_list2)
    return sum1 + sum2 
```

```python
def josephus_problem_without_modulo(n, k):
    people = list(range(1, n + 1)) 
    result = []
    index = 0 
    
    while people:
        index = index + k - 1
        # 인덱스가 리스트 길이를 초과할 때 조정
        while index >= len(people): 
            index = index - len(people)
        
        removed_person = people.pop(index)
        result.append(removed_person)
    
    return result      



n, k = map(int, input().split())
josephus_problem_without_modulo(n, k)
```
