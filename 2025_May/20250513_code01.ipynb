# 1. Cloud Architecture and Services

### 1. 글로벌 뼈대 (Global Backbone)

- **Azure Global Infrastructure (글로벌 인프라)**
    
    300 + 데이터센터와 Microsoft 전용 WAN으로 이뤄진 물리·네트워크 토대
    
- **Azure Global Network (Wide Area Network / 글로벌 네트워크)**
    
    60 + Region, 190 + Edge Site, 170 k mile 광섬유로 연결된 세계망
    
    - **Point of Presence (PoP / 접속 지점)** – 인터넷 트래픽이 Microsoft 망 안팎으로 드나드는 관문
    - **Regional Network Gateway (지역 네트워크 게이트웨이)** – 같은 Azure 지역 내 여러 데이터센터를 초고속 회선으로 묶어 주는 ‘허브’이자, 다른 지역 가상 네트워크와도 안전하게 연결해 주는 관문입니다.

---

### 2. 지리적 계층 (Geography → Region → Zone)

1. **Geography (지리)** – 국가·규제 범위 같은 최상위 묶음
2. **Region (지역)** – 가까운 DC들이 짧은 지연 네트워크로 묶인 단위
    - **Recommended Region (추천 지역)** – 서비스가 가장 많고 Zone 지원
    - **Alternate Region (대체 지역)** – DR·지연 최적화용, Zone 미지원
    - **Special Region (특수 지역)** – 美 정부·중국 21Vianet 등 규제 목적 격리 Region
    - **Region Pair (지역 쌍)** – 100 마일+ 떨어진 두 Region을 짝지어 데이터 복제·재해복구를 자동화
3. **Availability Zone (가용성 영역)** – 한 Region 안 서로 다른 건물·전력·네트워크를 가진 세 개 데이터센터. 한 Zone이 죽어도 서비스 지속 

---

### 3. 가용성 · 복원력 (Availability & Resiliency)

- **Availability Set (가용성 집합)** – 같은 데이터센터 안 VM을 장애 도메인·업데이트 도메인에 분산해 하드웨어 고장 대비
- **서비스 지역성 모드** (Zone을 어떻게 쓰느냐)
    - **Zonal Service (영역 서비스)** – 사용자가 특정 Zone 선택 배치
    - **Zone-redundant Service (영역 중복 서비스)** – 리소스를 Zone 간 자동 복제
    - **Always-available Service (항상 사용 가능 서비스)** – Zone 상관없이 모든 Region에서 제공
- **Disaster vs Incident (재해 vs 장애)** – 외부·통제불가 대형 사고 ↔ 내부·관리로 해결 가능 사고

---

### 4. 서비스 분류 (Service Classification)

| 계층 | 설명 | 지역 배포 패턴 |
| --- | --- | --- |
| **Foundational (기본)** | VM·Storage 등 핵심 IaaS/PaaS | 대부분 Region에 빠르게 배포 |
| **Mainstream (주류)** | 광범위 사용 서비스 | 추천 Region 우선·수요 따라 확대 |
| **Strategic (전략적)** | 특정 산업·전용 HW 기반 | 수요 기반 선택적 배포 |
| **Non-regional (비지역)** | 포털·Policy 등 관리 서비스 | 모든 Region 자동 제공 |

---

### 5. 엣지 · 위성 플랫폼 (Edge & Space)

- **Edge Computing (엣지 컴퓨팅)** – 데이터 원천 가까이서 처리해 지연↓
    - **Azure Edge Zone** – 통신사 5G/사내 MEC에 붙인 소형 Azure Region
- **Azure Space** – 위성·우주 데이터를 곧바로 Azure로 연결하는 에코시스템
    - **Azure Orbital Ground Station** – 지상국 as-a-Service, 위성이 보낸 데이터를 Azure로 수신

---

### 6. 요약 마인드맵 (글로 그려보기)

```
Global Backbone
 ├─ Physical DC / Azure WAN
 │   ├─ PoP
 │   └─ Regional Network Gateway
 └─ Geography
      └─ Region
           ├─ Zone (x3)
           │   ├─ Zonal Service
           │   ├─ Zone-redundant Service
           │   └─ Always-available Service
           ├─ Availability Set (inside DC)
           ├─ Region Pair
           ├─ Recommended / Alternate / Special
           └─ Service Categories
Edge & Space (별도 사이드)

```

## 1 네트워크 기본 토대

| 개념 | 쉬운 정의 |
| --- | --- |
| **컴퓨터 네트워크** | 여러 장치를 유선·무선으로 묶어 자료와 자원을 주고받는 구조 |
| **계층형 아키텍처(OSI 7계층)** | 복잡한 통신 과정을 물리–응용 7단계로 쪼개 관리하기 쉽게 만든 설계 |

| **라우팅** | 여러 라우터가 **패킷의 최적 길**을 계산해 다음 위치로 넘겨주는 과정 |
| **네트워크 프리픽스** | IP 주소 앞부분. ‘우리는 같은 네트워크야!’라고 표시하는 구간 |

# 3. 실습

### 왜 이 실습이 “AI 웹·앱 서비스” 에 도움 되는가

AI 모델을 실제 서비스에 넣으면 **예측 요청(request)** 이 몰릴 때가 많습니다.

- **두 대 이상 추론 서버**(예: GPU 가상 머신)로 부하를 분산해야 속도가 유지됩니다.
- **Azure Load Balancer**는 이런 다중 서버 앞에서 “자동 교통 정리”를 해 주는 장치입니다.
- 이번 PDF는 *“부하 분산이 되는 클라우드 인프라를 직접 손으로 만들어 보는 과정”* 을 다룹니다. 실습을 따라 하면 **① 안전한 네트워크 → ② 로드 밸런서 → ③ 서버 풀**까지 한 번에 갖춘 **기본 백엔드 틀**이 완성됩니다.
    
    ---
    
    ## 전체 흐름 한눈에 보기
    
    | 단계 | 해야 하는 이유 | 전문 용어 쉽게 설명 |
    | --- | --- | --- |
    | **1. 리소스 그룹 만들기** | 실습에서 쓰는 모든 리소스를 **폴더처럼 한데 묶어** 나중에 관리·삭제를 쉽게 함 | *Resource Group* – Azure 리소스의 **폴더** |
    | **2. NAT 게이트웨이 배치** | 가상 머신들이 **공용 인터넷으로 나갈 때** 고정 IP를 쓰게 하고, 인바운드 차단으로 보안을 높임 | *NAT Gateway* – 내부 IP ↔ 공용 IP를 **번역**해 주는 관문 |
    | **3. 가상 네트워크(VNet)와 서브넷 생성** | 추론 서버와 관리 장치를 **같은 사설 네트워크**에 담아 두고, 서브넷으로 구역을 나눠 트래픽을 정돈 | *Virtual Network* – 클라우드 속 **사설 LAN***Subnet* – 그 LAN 안의 **방** |
    | └ **Bastion Host 만들기** | 가상 머신에 **공용 IP를 노출하지 않고도** 웹 브라우저로 SSH‧RDP 접속 | *Bastion* – “원격 데스크톱 전용 문지기” |
    | **4. Azure Load Balancer 만들기** | 여러 VM 앞에 **단일 공용 IP(Front-end)** 를 세워 요청을 골고루 분산 | *Front-end IP* – 사용자에게 보이는 IP*Back-end Pool* – 트래픽을 받을 **서버 목록** |
    | └ **Health Probe·LB Rule 설정** | VM이 **정상인지 5초마다 체크** → 살아 있는 서버에만 80 포트 트래픽을 보냄 | *Health Probe* – “살아 있니?” 검사 신호*Load-balancing Rule* – “어떤 포트 요청을 어떻게 나눌까” 규칙 |
    | **5. 가상 머신 2대 배포 (각기 Zone 1, Zone 2)** | 서버를 서로 다른 데이터센터에 배치해 **하드웨어 장애에도 서비스 지속** | *Virtual Machine* – 클라우드 속 컴퓨터*Availability Zone* – 같은 지역 안 **물리적으로 다른 건물** |
    | └ **Network Security Group(NSG) 적용** | VM 직접 포트는 닫고 **Load Balancer → VM** 흐름만 허용해 보안 강화 | *NSG* – 클라우드 방화벽 규칙 세트 |
    | **6. IIS 웹 서버 설치** | VM마다 간단한 “Hello World” 페이지를 올려 **로드 밸런싱 동작 확인** | *IIS* – Windows 기본 웹(HTTP) 서버 |
    | **7. 브라우저 테스트** | 로드 밸런서 IP로 접속해 **새로 고침(F5)** 할 때마다 VM1, VM2가 번갈아 노출되는지 확인 |  |
    | **8. 리소스 그룹 한 번에 삭제** | 실습 리소스는 과금이 되므로 **정리** |  |
    
    > 실제 AI 서비스에서는 VM 대신 컨테이너 AKS 노드, “80 포트” 대신 /predict API 로 바뀔 뿐, 구조·개념은 그대로 사용됩니다.

> 비유:
> 
> - **Load Balancer** = 교차로 신호등, 손님(HTTP 요청)을 두 가게(VM)로 번갈아 보냄
> - **Bastion Host** = 경비실을 거쳐야만 내부로 들어오는 안전문
> - **NAT Gateway** = 집 안 사람이 밖으로 나갈 때 가명(공용 IP) 쓰게 해 주는 출구

---

### 구성 요소를 짧게 다시 정리

| 구성 | 하는 일 | 왜 필요한가? |
| --- | --- | --- |
| **Resource Group** | 실습 리소스 전체를 한 폴더처럼 묶음 | 나중에 통째로 삭제·비용 관리 |
| **Virtual Network (VNet)** | 퍼블릭 인터넷과 격리된 우리 전용 네트워크 | AI 서버를 안전하게 묶어 두기 |
| **Subnet** | VNet 안의 방 | 관리용 Bastion Subnet, 서비스용 Web Subnet 역할 구분 |
| **Bastion Host** | 브라우저에서 바로 SSH·RDP 접속 | VM에 공용 IP를 달지 않아도 원격 관리 |
| **NAT Gateway** | VM이 외부로 나갈 때 고정 공용 IP 부여 | 로그 추적·접근 제어 쉽게 하고 인바운드는 차단 |
| **Azure Load Balancer** | 한 개 공용 IP로 다수 VM에 트래픽 분산 | AI 예측 요청이 몰려도 속도 유지 |
| **Health Probe** | VM 살아 있나 5초마다 체크 | 고장 VM으로 요청 안 보냄 |
| **VM 1, VM 2 (다른 Zone)** | AI 모델·웹 API 실제 실행 | 물리적 장애에도 서비스 지속 |
| **Network Security Group(NSG)** | 포트 허용·차단 규칙 | 최소 권한 보안(Zero Trust) |

---

### 전체 흐름(순서)

1. **사용자**는 Load Balancer의 **공용 IP**로 `/predict` 같은 API 요청을 보낸다.
2. Load Balancer는 **Health Probe** 결과를 보고 살아 있는 VM에만 요청을 전달한다.
3. 선택된 **VM**이 AI 모델로 예측을 수행하고 응답을 Load Balancer経由로 사용자에게 돌려준다.
4. 운영자가 VM을 수정해야 할 땐 **Bastion Host** IP로 접속해 내부 네트워크로 안전하게 들어간다.
5. VM이 밖으로 업데이트 파일을 받는 등 인터넷으로 나갈 땐 **NAT Gateway**를 경유해 고정 공용 IP로 나간다.
