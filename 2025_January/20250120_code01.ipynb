```python
import json

# JSONL 파일 경로
file_path = r"C:\Users\1\Documents\GitHub\2024_AvocadoFarm\fetaqa.jsonl"

# JSONL 파일 읽기
def load_jsonl(file_path):
    data = []
    with open(file_path, "r", encoding="utf-8") as file:
        for line in file:
            data.append(json.loads(line.strip()))
    return data

# 데이터 출력
if __name__ == "__main__":
    data = load_jsonl(file_path)
    print(f"Loaded {len(data)} entries.")
    
    # 첫 번째 데이터 샘플 출력
    print("\nFirst entry:")
    print(json.dumps(data[0], indent=4, ensure_ascii=False))


```

```python
import json
import pandas as pd

# JSONL 파일 경로
file_path = r"C:\Users\1\Documents\GitHub\2024_AvocadoFarm\fetaqa.jsonl"

# JSONL 데이터 로드
with open(file_path, "r", encoding="utf-8") as file:
    data = [json.loads(line.strip()) for line in file]

# 첫 번째 항목 가져오기
first_entry = data[0]

# 'table_in_csv' 데이터를 DataFrame으로 변환
table_csv = first_entry["table_in_csv"]
df = pd.DataFrame([row.split(",") for row in table_csv.strip().split("\n")[1:]],
                  columns=table_csv.strip().split("\n")[0].split(","))

# DataFrame 출력
df
```
