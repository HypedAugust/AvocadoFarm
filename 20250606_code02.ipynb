```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.collections import LineCollection
from matplotlib.colors import LinearSegmentedColormap

plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.unicode_minus'] = False

# 1)  
labels = [
    "AI Literacy",
    "Game Domain Understanding",
    "Cross-functional Collaboration",
    "Team-level Strategy & Execution",
    "Global Communication",
    "Research / Publication Experience"
]
scores = [7.5, 9, 7.5, 7, 7.5, 6]
scores += scores[:1]                               # 

# 2)  
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]

# 3)   ()
cmap = LinearSegmentedColormap.from_list("purple_blue", ["purple", "blue"])
colors = cmap(np.linspace(0, 1, len(labels)))

# 4)  
fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(polar=True))

# 4-1)   ( )
ax.fill(angles, scores, color='cornflowerblue', alpha=0.25, zorder=1)

# 4-2)      
segments = []
for i in range(len(labels)):
    seg = [(angles[i], scores[i]), (angles[i + 1], scores[i + 1])]
    segments.append(seg)
line_collection = LineCollection(segments,
                                 colors=colors,
                                 linewidths=2,
                                 zorder=2)
ax.add_collection(line_collection)

# 5)  
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=11)

ax.set_yticks([2, 4, 6, 8, 10])
ax.set_yticklabels([''] * 5)          #   
ax.set_ylim(0, 10)

# 6) ()  ,   
ax.xaxis.grid(False)                  #  
ax.yaxis.grid(True, color='gray', linestyle='dotted', linewidth=0.6)

# 7)   
ax.set_title("Self-assessed Competency Radar for DL Strategy & Ops",
             size=14, pad=20)
ax.text(0, -1.6,
        " Scores based on expected max level of a 38Y associate",
        ha='center', va='center', fontsize=9)

plt.tight_layout()
plt.show()

```
