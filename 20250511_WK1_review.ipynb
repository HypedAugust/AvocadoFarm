Cloud Computing : 인터넷을 통해 온디맨드로 컴퓨팅 리소스를 제공하는 것을 의미. 스토리지, 서버, 데이터베이스, 소프트웨어, 네트워킹과 같은 다양한 서비스를 인터넷 상의 원격 서버를 통해 제공

---

Virtualization : 다양한 컴퓨팅 자원의 가상 버전을 생성함으로써 하나의 물리적인 하드웨어에서 여러 환경이 동시에 실행되어 소프트웨어를 하드웨어와 분리시키는 기술

- Server Virtualization : 물리 서버를 여러개의 가상머신 (VM)으로 분할 하여 각각 독립적인 OS, 응용 프로그램을 실햄. 서버 리소스를 효율적으로 활용.
- OS Virtualization :  Containerization 또는 Container화라고 부름.  하나의 운영 체제 커널을 공유하는 여러 격리된 사용자 공간 인스턴스인 컨테이너를 생성. 가벼우며 응용 프로그램을 빠르게 배포하게 함.
    - 커널 : OS의 핵심부분으로 소프트웨어 요청을 하드디스크에 저장하는 작업을 수행
- 네트워크 가상화 : 네트워크 가상화는 스위치, 라우터, 방화벽과 같은 네트워크 리소스를 추상화하고 가상화하는 것을 의미합니다. 가상 네트워크를 생성하여 네트워크 트래픽을 논리적으로 분리하고 보다 유연하고 확장 가능한 네트워크 구성을 가능하게 합니다.
    - 데이터 스냅샷 : 특정 시점의 데이터 상태를 사진찍듯이 저장하는 기능
    - 스위치 : 같은 네트워크 내에서 컴퓨터들을 연결해주는 장치
    - 라우터 : 서로다른 네트워크를 연결해주는 장치
    - 방화벽 : 네트워크 보안 장치

---

DaaS (Desktop as a Service) : DaaS는 기술적으로는 VDI에 기반을 두고 있습니다. VDI는 기업에서 자체적으로 서버를 운영하여 데스크탑 가상화를 이용한다면, DaaS는 클라우드 서비스 제공업체에서 사용자에게 VDI 서비스를 제공하고 종량제 방식으로 과금을 하는 서비스 입니다.

- VDI : Virtual Desktop Infrastructure : 회사 서버에 여러개의 가상 컴퓨터를 만들어 직원들이 원격으로 사용하는 기술

---

CI/CD : 소프트 웨어 개발 방법론. Continuous Integration and Continuous Delivery

DevOps : 개발과 운영을 함께하는 문화 철학. CI/CD를 포함하는 개념이다. 코드 변경사항을 자동으로 통합하고 배포하는 프로세스. (GIthub 쓰는 상태라 보면됨)

---

Cloud Computing 유형

- 클라우드 배포 모델 : Private, Public, Hybrid, Multi Cloud 모델로 나뉜다.
- 클라우드 서비스 유형 : Iaas, Paas, Saas로 분류된다.
    - IaaS – Infrastructure as a Service : Lift and Shift Migration 이나 Setup Test, Development 떄 사용하며, 예시로 Azure, Aws가 있다.
- PaaS – Platform as a Service : 개발 프로세스를 단순화하여 개발자가 인프라 관련 문제가 아닌 코딩과 애플리케이션 로직에 집중할 수 있도록 함. 예시로 각종 데스크톱앱을 떠올릴 수 있음. 포토샵 등등.
- SaaS – Software as a Service : 구독 기반으로 인터넷을 통해 소프트웨어 애플리케이션을 제공. 업데이트와 유지보수를 제공업체가 처리. 예시로 Gmail등이 있다.

---

Shared Responsibility Model : 공동 책임 모델

IT 인프라 모델의 비용 유형 : CapEx (Capital Expenditure), OpEX(Operational Expenditure)

- Consumption based Model : 클라우드 컴퓨팅은 OpEX에 속한다.

Cloud Computing의 특징 : High availability (고 가용성, HA), Scalability, Reliability, Predictability, Security, Governance, Manageability

- High availability : 고 가용성, 지속적으로 작동할 수 있도록 설계된 시스템

SLA : Service Level Agreement
서비스 제공 업체와 고객과 서비스 수준에 대해 협약을 맺는 것을 의미함.

Scale up (Vertical Scaling) : 하드웨어의 성능을 높여 스케일업 함.
Scale out( Horizontal Scaling ) : 하드웨어를 추가하여 스케일업 함.

Elasticity Computing : 자원을 동적으로 조정할 수 있는 능력

Governance: 클라우드 거버넌스는 조직 내에서 클라우드 사용을 제어하는 방법.

---

API : Application Programming Interface컴퓨터 프로그램들이 서로 대화하는 방식

- REST 기반 API : URL 주소를 통해 데이터를 요청하고 받는 방법의 API 방식
- HTTPS 기반 클라이언트 :  인터넷에서 정보를 안전하게 주고 받는 방법

CLI 환경:  Command Line Interface 키보드로만 컴퓨터를 조작하는 방법

Distro : Distribution의 약자로 Linux 운영체제의 다양한 버전을 말한다.

SSH (Secure SHeell) : 보안되지 않은 네트워크를 통해 컴퓨터 간의 안전한 통신을 가능하게 하는 암호화 네트워크 프로토콜. 주로 서버와 같은 원격 시스템에 안전하게 접근하고 관리하기 위해 명령 줄 인터페이스를 통해 사용됨.

DNS : Domain Name system 도메인명을 뜻한다.

Apache 서버 : 웹사이트를 인터넷에 보여주는 프로그램.

Certbot : 웹사이트에 보안 인증서를 자동으로 설치해주는 도구이다.

PHP : 웹사이트를 만드는 프로그래밍 언어 중 하나
SSL 인증서 :  인터넷에서 '안전한 자물쇠' 역할을 함.

VM의 구성요소 정의

1. NIC : Network Interface Card
2. NSG : Network Security Group
3. Public IP : 공용 IP 주소

---

빈 컴퓨터(VM) → 접속 도구(SSH) → 웹 서버(Apache) → 데이터베이스(MariaDB) → 프로그래밍 언어(PHP) → WordPress 설치 → 주소 설정(DNS) → 보안 설정(SSL)

- SSH 접속이유 : 원격으로 컴퓨터를 조작하려면 연결이 필요하기 때문
- apt update, upgrade : 최신 보안 패치, 버그 수정 받기
- Apache설치 이유 : 웹서버가 있어야 웹사이트를 보여줄 수 있기 떄문
- Maria DB :  데이터 베이스가 있어야 게시글, 사용자 정보 저장이 가능
- PHP 사용이유 :  wordpress가 php로 만들어졌고 wordpress의 요청을 apache가 받을 수 있어야 하기 떄문.
- certbot으로 SSL 인증서 설치 이유 : 웹사이트를 안전하게 만들기 위함.
