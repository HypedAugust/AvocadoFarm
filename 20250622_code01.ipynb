```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(-3,3,100) # linear space 처음부터 끝까지 100점으로 만들어라. 
y = x**2

plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-3,3,100) # linear space 처음부터 끝까지 100점으로 만들어라. 
y = np.exp(x)

plt.plot(x,y)
plt.show()
```
