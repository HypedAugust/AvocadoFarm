```python
quiz = ["3 - 4 = -3", "5 + 6 = 11"] #["X", "O"]
def solution(quiz):
    result = []
    for i in quiz:
        parts = i.split("=")
        left = parts[0]
        answer = int(parts[1].strip())

        if "+" in left:
            numbers = left.split("+")
            result_temp = int(numbers[0]) + int(numbers[1])
        elif "-" in left:
            numbers = left.split("-")
            result_temp = int(numbers[0]) - int(numbers[1])
        
        if result_temp == answer:
            result.append("O")
        else:
            result.append("X")
        
    return result 

print(solution(quiz))
```
